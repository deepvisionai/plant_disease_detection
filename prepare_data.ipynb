```python
ROOT_PATH = '../input/plant_doc_original_detection_dataset/PlantDoc-Object-Detection-Dataset'
TRAIN_PATH = f"{ROOT_PATH}/TRAIN"
TEST_PATH = f"{ROOT_PATH}/TEST"

train_images = glob.glob(f"{TRAIN_PATH}/*.jpg")
test_images = glob.glob(f"{TEST_PATH}/*.jpg")
train_annots = glob.glob(f"{TRAIN_PATH}/*.xml")
test_annots =glob.glob(f"{TEST_PATH}/*.xml")

train_images.sort()
test_images.sort()
train_annots.sort()
test_annots.sort()

# print(f"Number of training images: {len(train_images)}")
# print(f"Number of test images: {len(test_images)}")
# train_images.extend(test_images)
# all_images = train_images
# train_annots.extend(test_annots)
# all_annots = train_annots
# print(f"All images: {len(all_images)}")
# print(f"All annotations: {len(all_annots)}")

all_train_image_names = []
all_train_annot_names = []
all_test_image_names = []
all_test_annot_names = []

for i in tqdm(range(len(train_images))):
    image_name = train_images[i].split(os.path.sep)[-1].split('.')
    image_name = '.'.join(image_name[:-1])
    all_train_image_names.append(image_name)
for i in tqdm(range(len(test_images))):
    image_name = test_images[i].split(os.path.sep)[-1].split('.')
    image_name = '.'.join(image_name[:-1])
    all_test_image_names.append(image_name)
    
for i in tqdm(range(len(train_annots))):
    annot_name = train_annots[i].split(os.path.sep)[-1].split('.')
    annot_name = '.'.join(annot_name[:-1])
    all_train_annot_names.append(annot_name)
for i in tqdm(range(len(test_annots))):
    annot_name = test_annots[i].split(os.path.sep)[-1].split('.')
    annot_name = '.'.join(annot_name[:-1])
    all_test_annot_names.append(annot_name)
    
assert(len(all_train_annot_names) == len(train_annots))
assert(len(all_test_annot_names) == len(test_annots))
assert(len(all_train_image_names) == len(train_images))
assert(len(all_test_image_names) == len(test_images))

for name in tqdm(all_train_image_names):
    if name in all_train_annot_names:
        shutil.copy(f"{TRAIN_PATH}/{name}.jpg", f"{ROOT_PATH}/images")
        shutil.copy(f"{TRAIN_PATH}/{name}.xml", f"{ROOT_PATH}/annotations")
for name in tqdm(all_test_image_names):
    if name in all_test_annot_names:
        shutil.copy(f"{TEST_PATH}/{name}.jpg", f"{ROOT_PATH}/images")
        shutil.copy(f"{TEST_PATH}/{name}.xml", f"{ROOT_PATH}/annotations")
    

# for i in tqdm(train_images):
#     shutil.copy(i, f"{ROOT_PATH}/images")
# for i in tqdm(test_images):
#     shutil.copy(i, f"{ROOT_PATH}/images")
# for i in tqdm(train_annots):
#     shutil.copy(i, f"{ROOT_PATH}/annotations")
# for i in tqdm(test_annots):
#     shutil.copy(i, f"{ROOT_PATH}/annotations")
```

```python
COPY_FROM_ROOT_PATH = '../input/plant_doc_original_detection_dataset/PlantDoc-Object-Detection-Dataset'
ANNOT_PATH = f"{COPY_FROM_ROOT_PATH}/annotations"
IMAGE_PATH = f"{COPY_FROM_ROOT_PATH}/images"

COPY_TO_ROOT_PATH = 'data'

annotations = glob.glob(f"{ANNOT_PATH}/*.xml")
images = glob.glob(f"{IMAGE_PATH}/*.jpg")

annotations.sort()
images.sort()

print(annotations[:1])
print(images[:1])

print(f"Total number of training images: {len(images)}")
print(f"Total number of training annotations: {len(annotations)}")

annot_names = []
image_names = []
for annot_num, annot in enumerate(annotations):
    annot_name = annot.split(os.path.sep)[-1].split('.')[-2]
    annot_names.append(annot_name)
for image_num, image in enumerate(images):
    image_name = image.split(os.path.sep)[-1].split('.')[-2]
    image_names.append(image_name)
    
for name1, name2 in zip(image_names, annot_names):
    count1 = image_names.count(name1)
    count2 = annot_names.count(name2)
    if count1 > 1:
        print(f"Image {name1} present {count1} number of times")
    if count2 > 1:
        print(f"Annotation {name2} present {count2} number of times")
for name in annot_names:
    count = annot_names.count(name)
    if count > 1:
        print(f"Annotation {name} present {count} number of times")


# check for images without XML files
extra_image_counter = 0
for num, name in enumerate(image_names):
    if image_name not in annot_names:
        print(f"EXTRA IMAGE: {image_name}.jpg")
        extra_image_counter += 1
# check for XML files without images
extra_xml_counter = 0
for num, name in enumerate(annotations):
    if annot_name not in image_names:
        print(f"EXTRA XML: {annot_name}.xml")
        extra_xml_counter += 1
print(f"Total extra images: {extra_image_counter}")
print(f"Total extra XMLs: {extra_xml_counter}")

assert (len(annotations) == len(images)), 'Number of images and annotataion files do not match'

train_split = 80
valid_split = 10
test_split = 10

# set the train, evaluation/validation, and test ratio
train_ratio, valid_ratio, test_ratio = int((train_split/100)*len(images)), int((valid_split/100)*len(images)), int((test_split/100)*len(images))
print(f"Train data points: {train_ratio}")
print(f"Validation/evaluation data points: {valid_ratio}")
print(f"Test data points: {test_ratio}")

#############################################################
random.seed(42)
r = random.random() # randomly generating a real in [0,1)

# for training data
temp_annotations = glob.glob(f"{ANNOT_PATH}/*.xml")
temp_images = glob.glob(f"{IMAGE_PATH}/*.jpg")
temp_annotations.sort()
temp_images.sort()
random.shuffle(temp_annotations, lambda : r)  # lambda : r is an unary function which returns r
random.shuffle(temp_images, lambda : r) # using the same function as used in prev line so that shuffling order is same
for i in tqdm(range(0, train_ratio)):
    if i < 5:
        print(temp_annotations[i].split('.')[0].split(os.path.sep)[-1], temp_images[i].split('.')[0].split(os.path.sep)[-1])
    assert(temp_annotations[i].split('.')[0].split(os.path.sep)[-1] == temp_images[i].split('.')[0].split(os.path.sep)[-1]), 'Annotation and image do not match'
    shutil.copy(temp_annotations[i], f"{COPY_TO_ROOT_PATH}/train/annotations")
    shutil.copy(temp_images[i], f"{COPY_TO_ROOT_PATH}/train/images")

# for validation dataset
temp_annotations = glob.glob(f"{ANNOT_PATH}/*.xml")
temp_images = glob.glob(f"{IMAGE_PATH}/*.jpg")
temp_annotations.sort()
temp_images.sort()
random.shuffle(temp_annotations, lambda : r)  # lambda : r is an unary function which returns r
random.shuffle(temp_images, lambda : r) # using the same function as used in prev line so that shuffling order is same
for i in tqdm(range(train_ratio, train_ratio+valid_ratio)):
    if i < 5:
        print(temp_annotations[i].split('.')[0].split(os.path.sep)[-1], temp_images[i].split('.')[0].split(os.path.sep)[-1])
    assert(temp_annotations[i].split('.')[0].split(os.path.sep)[-1] == temp_images[i].split('.')[0].split(os.path.sep)[-1]), 'Annotation and image do not match'
    shutil.copy(temp_annotations[i], f"{COPY_TO_ROOT_PATH}/validation/annotations")
    shutil.copy(temp_images[i], f"{COPY_TO_ROOT_PATH}/validation/images")

# for test data
temp_annotations = glob.glob(f"{ANNOT_PATH}/*.xml")
temp_images = glob.glob(f"{IMAGE_PATH}/*.jpg")
temp_annotations.sort()
temp_images.sort()
random.shuffle(temp_annotations, lambda : r)  # lambda : r is an unary function which returns r
random.shuffle(temp_images, lambda : r) # using the same function as used in prev line so that shuffling order is same
for i in tqdm(range(train_ratio+valid_ratio, train_ratio+valid_ratio+test_ratio)):
    if i < 5:
        print(temp_annotations[i].split('.')[0].split(os.path.sep)[-1], temp_images[i].split('.')[0].split(os.path.sep)[-1])
    assert(temp_annotations[i].split('.')[0].split(os.path.sep)[-1] == temp_images[i].split('.')[0].split(os.path.sep)[-1]), 'Annotation and image do not match'
    shutil.copy(temp_annotations[i], f"{COPY_TO_ROOT_PATH}/test/annotations")
    shutil.copy(temp_images[i], f"{COPY_TO_ROOT_PATH}/test/images")
```

```python
def xml_to_csv(path):
    annotations_list = glob.glob(path + '/*.xml')
    annotations_list.sort()
    xml_list = []
    for xml_file in annotations_list:
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            # print('MEMBER', member)
            try:
                value = (root.find('filename').text,
                         int(root.find('size')[0].text),
                         int(root.find('size')[1].text),
                         member[0].text,
                         int(member[4][0].text),
                         int(member[4][1].text),
                         int(member[4][2].text),
                         int(member[4][3].text)
                         )
            except:
                value = (root.find('filename').text,
                         int(root.find('size')[0].text),
                         int(root.find('size')[1].text),
                         member[0].text,
                         int(member[5][0].text),
                         int(member[5][1].text),
                         int(member[5][2].text),
                         int(member[5][3].text)
                         )
            xml_list.append(value)
    column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df


def main():
    ROOT_PATH = 'data'
    for folder in [
        f"{ROOT_PATH}/train", 
        f"{ROOT_PATH}/validation",
        f"{ROOT_PATH}/test"]:
        xml_df = xml_to_csv(folder+'/'+'annotations')
        xml_df.to_csv((folder + '/' + 'labels/' + 'labels.csv'), index=None)
        print('Successfully converted xml to csv.')

main()
```

```python
"""
Sometimes the xmin, ymin, xmax, and ymax annotation values may be 
larger than the actual image dimension due to labelling error.
This might cause unkown errors down the line. This Python script
removes such erroneous rows from the CSV files and rewrites them 
to the disk.
"""

# ROOT_PATH = '/content'

print('INFO: In test CSV')
test_df = pd.read_csv(f"final_labels/final_test_csv.csv")
drop_indices = []

for i in range(len(test_df)):
    w, h = test_df['width'][i], test_df['height'][i]
    if test_df['xmin'][i] > w or test_df['xmax'][i] > w or test_df['ymin'][i] > h or test_df['ymax'][i] > h:
        print(test_df['filename'][i])
        drop_indices.append(test_df.index[i])
print(f"Drop indices: {drop_indices}")

test_df.drop(drop_indices, inplace=True)
# REWRITE THE CSV FILE TO DISK. VERY IMPORTANT
test_df.to_csv(f"final_labels/final_test_csv.csv", index=False)

##########################################################################

print('INFO: In valid CSV')
valid_df = pd.read_csv(f"final_labels/final_valid_csv.csv")
drop_indices = []

for i in range(len(valid_df)):
    w, h = valid_df['width'][i], valid_df['height'][i]
    if valid_df['xmin'][i] > w or valid_df['xmax'][i] > w or valid_df['ymin'][i] > h or valid_df['ymax'][i] > h:
        print(valid_df['filename'][i])
        drop_indices.append(valid_df.index[i])
print(f"Drop indices: {drop_indices}")

valid_df.drop(drop_indices, inplace=True)
# REWRITE THE CSV FILE TO DISK. VERY IMPORTANT
valid_df.to_csv(f"final_labels/final_valid_csv.csv", index=False)
##########################################################################

print('INFO: In training CSV')
train_df = pd.read_csv(f"final_labels/final_train_csv.csv")
drop_indices = []

for i in range(len(train_df)):
    w, h = train_df['width'][i], train_df['height'][i]
    if train_df['xmin'][i] > w or train_df['xmax'][i] > w or train_df['ymin'][i] > h or train_df['ymax'][i] > h:
        print(train_df['filename'][i])
        drop_indices.append(train_df.index[i])
print(f"Drop indices: {drop_indices}")

train_df.drop(drop_indices, inplace=True)
# REWRITE THE CSV FILE TO DISK. VERY IMPORTANT
train_df.to_csv(f"final_labels/final_train_csv.csv", index=False)
##########################################################################
```
